# Question 1: 
1. Create a line plot that displays the rolling mean CO values for one station with a one-week window

```python
#Value checks for date
df['Date'].value_counts()
```

```python
#Graph
px.scatter(df,
           x='Date',
           y='Value')
```

```python
#Station Value Counts
df['Station'].value_counts()
```

```python
#Selecting a single station 
newdf = df[df['Station'] == 'Tsawwassen'].reset_index(drop=True)
newdf
```

```python
#Value counts for newdf by Date
newdf['Date'].value_counts()
```

```python
#Single Time Series Test
px.line(newdf,
        x='Date',
        y='Value')
```

```python
#Reshaping and cleaning up
newdf = (newdf.pivot(index='Date',columns='Station')['Value']
      .reset_index()
      .rename_axis(None, axis=1)
      .fillna(0)
     )
newdf
```

```python
newdf = newdf.set_index('Date')
```

```python
#Get the mean of a week 
tmp = (newdf.rolling(168).mean()
              .dropna()
              .reset_index()
             )
tmp
```

```python
#Get further details of the new data frame
newdf.head()
```

```python
#Melting the data frame
tmp.melt(id_vars='Date',value_vars=['Tsawwassen'],var_name='Station',value_name='CO')
```

```python
#Printing the data frame date min and max 
print(tmp['Date'].min(),tmp['Date'].max())
```

```python
#Plotting the 7 day average values of station Tsawwassen
px.line(tmp,
        x='Date',
        y='Tsawwassen',
       width=800,
       height=600,
       title='CO 7 Day Average Values of Station Tsawwassen')
```

# Question 2:
2. Create a faceted line plot that displays the rolling mean CO values for each month of the sixmonth record. Use a facet_col_wrap argument of 2 to display the result as two columns and three rows.

```python
#Get the months that apply to the data frame
tmp['Month'] = tmp['Date'].apply(lambda x:x.split('-')[1]) 
tmp                              
```

```python
#Final Plot 
fig = px.line(tmp,
              x='Date',
              y='Tsawwassen',
              labels=dict(Date="Date By Month", Tsawwassen="CO"),
              title='CO Values of 6 Months',
              facet_col='Month',
              facet_col_wrap=2)

fig.show()
```

# Question 3:
3. Create a line plot displaying the same information using the color argument. 

```python
#Plot connected without using the facet arguments but using color alone
fig = px.line(tmp,
              x='Date',
              y='Tsawwassen',
              labels=dict(Date="Date By Month", Tsawwassen="CO"),
              title='CO Values of 6 Months',
              color='Month')

fig.show()
```

```python
#Final plot with facet arguments and color
fig = px.line(tmp,
             x='Date',
             y='Tsawwassen',
             labels=dict(Date="Date By Month", Tsawwassen="CO"),
             title='CO Values of 6 Months',
             color='Month',
              facet_col='Month',
              facet_col_wrap=2)
fig.show()
```

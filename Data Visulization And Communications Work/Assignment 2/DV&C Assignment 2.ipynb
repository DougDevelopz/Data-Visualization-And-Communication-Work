### Bar Chart

**What is the number of parking spaces avaliable based off of active building statuses?**

**To create this chart, I will use active building spaces to get the right number of parking spaces avaliable**

```python
#Final plot
fig = px.bar(df,
             x='Bldg Status',
             color='Total Parking Spaces',
             )

fig.show()
```

```python
#Counting total parking spaces
tmp = df['Bldg Status'].value_counts().to_frame().reset_index().rename(columns={'Bldg Status':'Total Parking Spaces','index':'Bldg Status'})
tmp
```

```python
# Final plot
fig = px.bar(tmp,
             x='Bldg Status',
             y='Total Parking Spaces',
             color='Bldg Status',
             color_discrete_sequence=px.colors.qualitative.D3,
             title='Parking Spaces Avaliable Based Off Building Status',
             width=800,
             height=600)

fig.show()
```

### Violin Chart

**What is the distrbution of parking spaces avaliable based off property types?**

**To create this chart, I will use property type to get the distribution of total parking spaces**

```python
# Final plot
fig = px.violin(prop_type,
             y='Total Parking Spaces',
             x='Property Type',
             color='Property Type',
             title='Distribution of Total Parking Spaces for Buildings Only',
             height=600,
             width=800
             )

fig.show()
```

### Heat Map
 **What is the number of Owned / Leased buildings by state with building property only?**
 
 **To create this chart I will use property types to look at the number of owned / leased buildings by state**

```python
#Cross datapaths 
tmp = pd.crosstab(prop_type['Bldg State'],prop_type['Owned/Leased'])
tmp
```

```python
#Final Plot
fig = px.imshow(tmp,
                title='Building Property Type By Building State and Owned / Leased',
                width=800,
                height=600)

fig.show()
```

### Box Plot

**What is the number of owned property based off building property types only?**

**To create this chart I will look at owned property based off different property types**

```python
# Get owned typed only.
owned = df[df['Owned/Leased']=='OWNED'].reset_index(drop=True)
owned
```

```python
#Final plot
fig = px.box(owned,
                 x='Owned/Leased',
                 y='Total Parking Spaces',
                 color='Property Type',
                 title='Owned Property With Different Property Types',
                 width=800,
                 height=600)
                 
fig.show()
```

## Histogram

**What is the number of parking spaces avaliable based off leased property?**

**To create this chart I will look at leased property based off different property types**

```python
# Get leased typed only.
leased = df[df['Owned/Leased']=='LEASED'].reset_index(drop=True)
leased
```

```python
#Final Plot
fig = px.histogram(leased,
                 y='Owned/Leased',
                 x='Total Parking Spaces',
                 color='Property Type',
                 title='Leased Property With Different Property Types',
                 width=800,
                 height=600)
                 
fig.show()
```

### Area Graph

**What is the the distribution of parking spaces based off states?**

**To create this chart I will look at the distribution of parking spaces based off states**

```python
#Final plot
fig = px.area(df,
                 x='Bldg State',
                 y='Total Parking Spaces',
                 color='Bldg State',
                 title='Distribution of Parking Spaces By State',
                 width=800,
                 height=600)
fig.show()
```

### Pie Chart

**What is the proportion of male to female smokers?**

**To create this chart I will look at the proportion of male to female smokers**

```python
# Actual number count of male to female smokers
gender = df2['sex'].value_counts()
gender
```

```python
#Final Plot but in percentages
fig = px.pie(gender,
           values=gender.value_counts(),
           names=gender.index,
           width=400,
           height=400,
           title='Proportion of Male To Female Smokers') 
fig.show()
```

### Scatter Plot

**What is the overall rating in BMI of male to female smokers?**

**To create this chart I will look at the overall BMI of male to female smokers**

```python
# BMI value counts
bmi = df2['bmi'].value_counts().to_frame().reset_index().rename(columns={'bmi':'sex','index':'bmi'})
bmi
```

```python
# Final plot
fig = px.scatter(bmi,
                x='bmi',
                y='sex',
                color='bmi',
                title='Overall Rating In BMI Based Off Male to Female Smokers',
                width=800,
                height=600)
fig.show()
```

### Sunburst

**What value of children are smokers based off id?**

**To create this chat I will look at the number of children that are smokes based by ID**

```python
# Getting the value count of children that are smokers
children = df2['children'].value_counts().to_frame().reset_index().rename(columns={'children':'smoker','index':'children'})
children
```

```python
# Final Plot
px.sunburst(children,
            path=['children','smoker'],
           title='Visual Of Children That Are Smokers By ID',
           width=800,
           height=600)
```

### Parallel Category Chat

**What's the BMI of these smokers / non smokers based off region?**

**To create this chart I will look at the BMI of smokers / non smokers based off region, but I'll reduce it to where BMI must be >= 40 to make the chart shorter**

```python
# Final Plot
px.parallel_categories(reduce,
            dimensions=['bmi','smoker','region'],
                      color='bmi',
                      width=1200,
                      height=800,
                      title='BMI of Smokers to Non Smokers By Region')
```

### Tree Map

**What's the BMI of smokers to non smokers based off age?**

**To create this chart I will look @ the BMI of smokers to non smokers based off their age

```python
# Final Plot
px.treemap(reduce,
           path=['smoker','bmi','age'],
          color='bmi',
          width=1200,
          height=800,
          title='BMI of Smokers to Non Smokers By Age')
```

### Strip Chart

**What are the charges from smokers to non smokers?**

**To create this chart I will look @ the charges based off of smokers to non smokers**

```python
# Final plot
fig = px.strip(df2,
              x='charges',
              y='smoker',
              color='smoker',
              width=800,
              height=600,
              title='Charges From Smokers to Non Smokers')
fig.show()
```
